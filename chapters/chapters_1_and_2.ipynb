```python
# Malaria Case Study (2.3.1)
malaria_data = pd.DataFrame({'treatment': ['vaccine', 'placebo'], 'infection': [5, 6], 'no infection': [9, 0]})
malaria_data.head()
```

```python
# Cmpute number of infected and not infected patients
num_infected = malaria_data['infection'].sum()
num_not_infected = malaria_data['no infection'].sum()
num_patients = num_infected + num_not_infected
num_vaccine = malaria_data['infection'][0] + malaria_data['no infection'][0]
num_placebo = malaria_data['infection'][1] + malaria_data['no infection'][1]
probability_infected_vaccine = malaria_data['infection'][0] / (malaria_data['infection'][0] + malaria_data['no infection'][0])
probability_infected_placebo = malaria_data['infection'][1] / (malaria_data['infection'][1] + malaria_data['no infection'][1])
probability_difference = probability_infected_placebo - probability_infected_vaccine
print(f'Probability of infection with vaccine: {probability_infected_vaccine:.2f}')
print(f'Probability of infection with placebo: {probability_infected_placebo:.2f}')
```

```python
# Randomize trial (independence model)
trials = int(1e6)
infections = np.array([1]*num_infected + [0]*num_not_infected)
trial_results = np.empty(trials)
for i in range(trials):
    np.random.shuffle(infections)
    infected_vaccine = infections[:num_vaccine].sum()
    infected_placebo = infections[num_vaccine:].sum()
    probability_infected_vaccine_randomized = infected_vaccine / num_vaccine
    probability_infected_placebo_randomized = infected_placebo / num_placebo
    trial_results[i] = probability_infected_placebo_randomized - probability_infected_vaccine_randomized
```

```python
# Plot histogram of randomized trial results
plt.figure(figsize=(10, 6))
plt.hist(trial_results, bins=20, color='purple', alpha=0.7, weights=100*np.ones(len(trial_results)) / len(trial_results))
plt.title('Histogram of Randomized Trial Results')
plt.xlabel('Probability Difference (Placebo - Vaccine)')
plt.ylabel('Frequency (%)')
plt.grid()
plt.show()
```

```python
# Calculate the probability of observing a difference as extreme as the one observed
p_value = np.mean(trial_results >= probability_difference)
print(f'Probability of observing a difference as extreme as {probability_difference:.2f}: {p_value:.4f}')
```

```python
# Homework plot generation, four plots with different associations

# Generate data for the plots
x1 = np.linspace(0, 10, 100)
y1 = 2 * x1 + np.random.normal(0, 3, size=100)  # Positive linear association

x2 = np.linspace(0, 10, 100)
y2 = -1 * x2 + np.random.normal(0, 2, size=100)  # Negative linear association

x3 = np.linspace(0, 10, 100)
y3 = x3**2 + np.random.normal(0, 10, size=100)  # Non-linear positive association

x4 = np.random.uniform(0, 10, 100)
y4 = np.random.uniform(0, 10, 100)  # Independent variables

# Create the plots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Positive linear association
axs[0, 0].scatter(x1, y1, color='blue', alpha=0.7)
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')
axs[0, 0].grid()

# Plot 2: Negative linear association
axs[1, 0].scatter(x2, y2, color='red', alpha=0.7)
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')
axs[1, 0].grid()

# Plot 3: Non-linear positive association
axs[1, 1].scatter(x3, y3, color='green', alpha=0.7)
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Y')
axs[1, 1].grid()

# Plot 4: Independent variables
axs[0, 1].scatter(x4, y4, color='purple', alpha=0.7)
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')
axs[0, 1].grid()

plt.tight_layout()
plt.show()
```
